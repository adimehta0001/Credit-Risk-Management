```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import pickle
np.random.seed(42)
n_samples = 500

data = {
    'age': np.random.randint(18, 75, n_samples),
    'credit_amount': np.random.randint(500, 15000, n_samples),
    'duration': np.random.randint(6, 72, n_samples)
}
df = pd.DataFrame(data)
df['Risk'] = np.where(
    (df['age'] > 30) & (df['credit_amount'] < 8000) & (df['duration'] < 36), 
    'good', 
    'bad'
)
X = df[['age', 'credit_amount', 'duration']]
y = df['Risk']
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)
with open('credit_model.pkl', 'wb') as f:
    pickle.dump(model, f)

print("SUCCESS: Synthetic Model Created. 'credit_model.pkl' is ready.")
```
